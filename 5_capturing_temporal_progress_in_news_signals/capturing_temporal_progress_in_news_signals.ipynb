```python
import torch
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    pipeline,
)

model_name="meta-llama/Llama-2-7b-chat-hf" # Model from hugging face
device_map = {"": 0} # Load the entire model on the GPU 0
use_4bit = True # Activate 4-bit precision base model loading
bnb_4bit_compute_dtype = "float16" # Compute dtype for 4-bit base models
bnb_4bit_quant_type = "nf4" # Quantization type (fp4 or nf4)
use_nested_quant = False # Activate nested quantization for 4-bit base models (double quantization)
compute_dtype = getattr(torch, bnb_4bit_compute_dtype) # Load tokenizer and model with QLoRA configuration

bnb_config = BitsAndBytesConfig(
    load_in_4bit=use_4bit,
    bnb_4bit_quant_type=bnb_4bit_quant_type,
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=use_nested_quant,
    )

# Check GPU compatibility with bfloat16
if compute_dtype == torch.float16 and use_4bit:
    major, _ = torch.cuda.get_device_capability()
    if major >= 8:
        print("=" * 80)
        print("Your GPU supports bfloat16: accelerate training with bf16=True")
        print("=" * 80)

# Load base model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map=device_map
)
model.config.use_cache = False
model.config.pretraining_tp = 1

# Load LLaMA tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right" # Fix weird overflow issue with fp16 training
```

```python
#generate_text = pipeline(task="text-generation", model=model, tokenizer=tokenizer, device_map="auto", torch_dtype=torch.bfloat16, max_length=256)
generate_text = pipeline(task="text-generation", model=model, tokenizer=tokenizer, device_map="auto", torch_dtype=torch.bfloat16)
```

```python
# template of prompt to be used

system_prompt = """
<s>[INST] <<SYS>>
You are an efficient analyzer of news texts. Always answer very very concisely. Never beat about the bush. No redundant words please. If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information.
<</SYS>>
"""

instruction_prompt = """
Shared here are two things - content of a news article and an event description. From the news article, spot sentences that announce or discuss or report subjects significantly similar to the event description shared. In your response, only share the relevant sentences from the news article and nothing else. No other text in your response please. If you think, the news article does not talk about anything at all similar to the event description, just reply with 'NOT APPLICABLE'.
"""

eoi_token = "[/INST]"
```

Now performing demonstration for splitting 20 news items on top signals

```python
for event in events_df['event']: # adding event columns in the dataframe
  news_df[event] = ""

news_df_archive = news_df.copy(deep=True)

news_df = news_df[:20] # trimming the dataset for computation timing; following below is a demo
```

```python
n = 4 # number of top signals to be considered for every news

for index, row in news_df.iterrows(): # this loop iterates over every row of the news_df
  news = row['news_content']
  for i in range (1, n + 1): # this loop iterates over the top 4 signals for every row
    signal_ = "signal_" + str(i)
    event_ = row[signal_][:-4] # removing the "_cos" string from signal name
    event_description = events_df[events_df['event'] == event_]['description'].iloc[0]
    prompt = system_prompt + " " + instruction_prompt + "NEWS ARTICLE: " + news + " \n" + "EVENT DESCRIPTION: " + event_description + "\n" + eoi_token + "\n"
    res = generate_text(prompt)[0]["generated_text"]
    news_df.at[index, event_] = res.split(eoi_token)[1] # cleaning model response for redundant text
```

### Now performing demonstration for splitting news items of any company of choice on top signals

```python
company_of_choice = "Commerzbank"

# filling NaN values with an empty string or any other value
news_df_archive["companies_in_news"] = news_df_archive["companies_in_news"].fillna("")

news_df_company = news_df_archive[news_df_archive["companies_in_news"].str.contains(company_of_choice)]

print(f"number of news items for {company_of_choice} = {len(news_df_company)}")

news_df_company.head(2)
```

```python
n = 4 # number of top signals to be considered for every news

for index, row in news_df_company.iterrows(): # this loop iterates over every row of the news_df_company
  news = row['news_content']
  for i in range (1, n + 1): # this loop iterates over the top 4 signals for every row
    signal_ = "signal_" + str(i)
    event_ = row[signal_][:-4] # removing the "_cos" string from signal name
    event_description = events_df[events_df['event'] == event_]['description'].iloc[0]
    prompt = system_prompt + " " + instruction_prompt + "NEWS ARTICLE: " + news + " \n" + "EVENT DESCRIPTION: " + event_description + "\n" + eoi_token + "\n"
    res = generate_text(prompt)[0]["generated_text"]
    news_df_company.at[index, event_] = res.split(eoi_token)[1] # cleaning model response for redundant text
```

### Combining news items within 'each signal' to generate searchable context, for every company of interest

```python
# new data frame for generating signal-wise context for companies of choice within particular dates
signalwise_news_context_df = pd.DataFrame(columns = ['company', 'news_start_timestamp', 'news_end_timestamp'])

for event_ in events_df['event']: # adding event columns in the dataframe
  signalwise_news_context_df[event_] = ""

signalwise_news_context_df
```

```python
signalwise_news_context_df.loc[0, "company"] = company_of_choice
signalwise_news_context_df.loc[0, "news_start_timestamp"] = min(news_df_company.news_date)
signalwise_news_context_df.loc[0, "news_end_timestamp"] = max(news_df_company.news_date)
signalwise_news_context_df
```

```python
for index, row in signalwise_news_context_df.iterrows():
  for event_ in events_df['event']:
    i = 1
    context = ""
    for item in news_df_company[event_]:
      if item != "":
        context = context + "(" + str(i) + ") " + item + " ***** "
        i = i + 1
    signalwise_news_context_df.at[index, event_] = context

signalwise_news_context_df   # this dataframe finally has "signal-wise context for companies of choice within particular dates"
```

### Prompting LLM to extract values from these signal-based contexts

```python
prefilled_columns_to_copy = ['company', 'news_start_timestamp', 'news_end_timestamp']

signalwise_news_value_df = signalwise_news_context_df[prefilled_columns_to_copy].copy(deep=True) # dataframe with llm's responses on the signal-based contexts

for event_ in events_df['event']: # adding event columns in the dataframe
  signalwise_news_value_df[event_] = ""

signalwise_news_value_df
```

```python
# template of prompt to be used here

system_prompt = """
<s>[INST] <<SYS>>
You are an efficient analyzer of news contexts. Always answer very very concisely. Never beat about the bush. No redundant words please. If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information.
<</SYS>>
"""

# instruction_prompt = """
# Shared here are two things - content of a news article and an event description. From the news article, spot sentences that announce or discuss or report subjects significantly similar to the event description shared. In your response, only share the relevant sentences from the news article and nothing else. No other text in your response please. If you think, the news article does not talk about anything at all similar to the event description, just reply with 'NOT APPLICABLE'.
# """

eoi_token = "[/INST]"
```

```python
for index, row in signalwise_news_value_df.iterrows():
    for event_ in events_df['event']:
        context = signalwise_news_context_df[event_].iloc[index]
        instruction_prompt = (
            "News discussing " + event_ + " are described as follows: " + events_df[events_df['event'] == event_]['description'].iloc[0] +
            " I am sharing a chain of news items about " + row['company'] + ". These items discuss dynamics, updates, movements, strategies, etc about " +
            row['company'] + " in the context of " + event_ + ". Based on the text shared below, please summarize the dynamics of the company in 4 to 5 sentences. " +
            "Do not miss out on any numbers or crucial details mentioned. Please also give a general sentiment of the news text around " + row['company'] + ". " +
            "The ask is to highlight (based on the text provided) what major things are happening in the world around " + row['company'] + " in general. Following " +
            "is the collection of news items: " + "\n" + context
        )
        prompt = system_prompt + "\n" +  instruction_prompt + eoi_token + "\n"
        res = generate_text(prompt)[0]["generated_text"]
        signalwise_news_value_df.at[index, event_] = res.split(eoi_token)[1]
```
